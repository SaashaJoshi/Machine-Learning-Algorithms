## Train - Test Split ##
### Advantages:###
    1. Algorithm evaluation technique is fast.
    2. Ideal for large datasets.
### Limitations:###
    1. It can have High Variance.
    2. It has Low Bias with larger datasets.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.csv"
col_names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
```

```python
df = pd.read_csv(url, names = col_names)
```

```python
df[:5]
```

```python
array = df.values
```

```python
X = array[:, :8]
Y = array[:, -1]
```

```python
# Y[:5]
```

```python
seed = 100
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = seed)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
result = model.score(X_test, Y_test)
accuracy = result*100.0
```

```python
print('Accurcy: ', accuracy)
# print(f'Accuracy: {accuracy}')
```

```python
import matplotlib.pyplot as plt

plt.title('Test Plot')
plt.scatter(X_test[:, 0], Y_test, color = 'Orange')
plt.plot(X_train[:, 0], model.predict(X_train))
plt.show()
```

```python
plt.title('Train Plot')
plt.scatter(X_train[:, 0], Y_train, color = 'Orange')
plt.plot(X_train[:, 0], model.predict(X_train))
plt.show()
```
